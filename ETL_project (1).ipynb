```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
dognames = "Resources/dognames.csv"
nationalnames = "Resources/nationalnames.csv"
statenames = "Resources/statenames.csv"
```

```python
dognames_df = pd.read_csv(dognames)
dognames_df = dognames_df.rename(columns={"Row_Labels":"Name","Count_AnimalName":"Count"})
dognames_df['Name']=dognames_df.Name.str.title()
dognames_df.set_index("Name",inplace=True)
dognames_df
```

```python
nationalnames_df=pd.read_csv(nationalnames)
nationalnames_df.head()
```

```python
statenames_df=pd.read_csv(statenames)
statenames_df.head()
```

```python
NY_df=statenames_df[(statenames_df['State']=='NY')]
df_2011=NY_df[(NY_df['Year']>=2011)]
df_2011.set_index("Id",inplace=True)
babynames_df=df_2011.drop(['Year'], axis=1)
babynames_df.head()
```

```python
merged = pd.merge(babynames_df,dognames_df, on=['Name','Count'])
new_df=merged.drop_duplicates(subset='Name', keep='first', inplace=False)
new_df.set_index("Name",inplace=True)
new_df.head()
```

```python
connection_string = "root:snoopy5@127.0.0.1/names_db"
engine = create_engine(f'mysql://{connection_string}')
```
